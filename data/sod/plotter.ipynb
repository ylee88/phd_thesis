```python
import numpy as np
import matplotlib.pyplot as plt
import slugger as slug

params = {'text.usetex': False, 'mathtext.fontset': 'stix'}
plt.rcParams.update(params)
%matplotlib inline
```

```python
# ref = slug.data1d.ascii('slug_sod_rk3_1024_10001.dat')

rk3_1024 = slug.data1d.ascii('slug_sod_rk3_1024_10001.dat')
pif_1024 = slug.data1d.ascii('slug_sod_pif_1024_10001.dat')
sf3_1024 = slug.data1d.ascii('slug_sod_sf3_1024_10001.dat')

# rk3 = slug.data1d.ascii('slug_sod_rk3_128_10001.dat')
# pif = slug.data1d.ascii('slug_sod_pif_128_10001.dat')
# sf3 = slug.data1d.ascii('slug_sod_sf3_128_10001.dat')

rk3 = slug.data1d.ascii('slug_sod_rk3_256_10001.dat')
pif = slug.data1d.ascii('slug_sod_pif_256_10001.dat')
sf3 = slug.data1d.ascii('slug_sod_sf3_256_10001.dat')

sf3_weno = slug.data1d.ascii('slug_sod_sf3_weno-diff_256_10001.dat')
```

```python
fig = plt.figure(figsize=(4,3), dpi=400)
ax = fig.add_subplot(1,1,1)

ref = rk3_1024

ax.plot(ref.x, ref.dens, 'g-', markersize=4, linewidth=0.5)
ax.plot(rk3.x, rk3.dens, 'g.', markersize=4, linewidth=0.5, label='RK3, '+r'$\rho$')

ax.plot(ref.x, ref.velx, 'b-', markersize=4, linewidth=0.5)
ax.plot(rk3.x, rk3.velx, 'b.', markersize=4, linewidth=0.5, label='RK3, '+r'$u$')

ax.plot(ref.x, ref.pres, 'r-', markersize=4, linewidth=0.5)
ax.plot(rk3.x, rk3.pres, 'r.', markersize=4, linewidth=0.5, label='RK3, '+r'$p$')

# ax.set(xlim=(4.0, 7.0))
ax.set(xlim=(0.0, 1.0))

# zoomed
axins = ax.inset_axes([0.05, 0.4, 0.25, 0.25])
axins.plot(ref.x, ref.velx, 'b-', markersize=4, linewidth=0.5)
axins.plot(rk3.x, rk3.velx, 'b.', markersize=4, linewidth=0.5)
axins.set(xlim=(0.8, 0.87), ylim=(0.87, 0.97))
axins.set_xticklabels('')
axins.set_yticklabels('')
axins.set_xticks([])
axins.set_yticks([])
ax.indicate_inset_zoom(axins, label='')
###

legend = ax.legend(fontsize=8, numpoints=3, fancybox=False, edgecolor='k', loc='lower center')
legend.get_frame().set_linewidth(.7)

ax.tick_params(axis='both', which='major', labelsize=7)
ax.set_title('Sod shock tube problem with RK3, ' + r'$t = 0.2$', fontsize=9)

ax.set_xlabel(r'$x$' + '-axis')
ax.set_ylabel(r'$\rho$, $u$, $p$')




fig.savefig('./sod_rk3.pdf', bbox_inches='tight', pad_inches=0)
```
