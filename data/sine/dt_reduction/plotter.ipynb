```python
import numpy as np
import matplotlib.pyplot as plt
import slugger as slug
import pandas as pd

params = {'text.usetex': False, 'mathtext.fontset': 'stix'}
plt.rcParams.update(params)
%matplotlib inline
```

```python
rk3 = pd.read_csv('./rk3.csv', index_col=0)
sf3 = pd.read_csv('./sf3.csv', index_col=0)
pif = pd.read_csv('./pif.csv', index_col=0)



NN = np.array(rk3.index)

a = np.float64(sf3.index[3])
b = sf3["dens"][a]
```

```python
fig = plt.figure(figsize=(4,3), dpi=400)
ax = fig.add_subplot(1,1,1)

a = np.float64(sf3.index[4])
b = sf3["dens"][a]
ax.plot(NN, (b*a**5)*(NN)**(-5.), 'k-.', linewidth=0.4)
ax.annotate(r'$\mathcal{O} (\Delta x^{5})$', xy=(9.5E1,3.E-7))

a = np.float64(sf3.index[1])
b = sf3["dens"][a]
# ax.plot(NN, (b*a**4)*(NN)**(-4.), 'm-.', linewidth=0.4)
# ax.annotate(r'$\mathcal{O} (\Delta x^{4})$', xy=(3.E1,2.E-2), color='m')

ax.plot(rk3['dens'], 'g.', markersize=4, linewidth=0.5, label='RK3')
ax.plot(pif['dens'], 'bx', markersize=4, linewidth=0.5, label='PIF')
ax.plot(sf3['dens'], 'r+', markersize=4, linewidth=0.5, label='SF-PIF')


ax.loglog()

ax.set_xlim(1.E1, 1.5E3)
ax.set_ylim(1.E-13, 5.E-2)

legend = ax.legend(fontsize=7, numpoints=1, fancybox=False, edgecolor='k')
legend.get_frame().set_linewidth(.7)

ax.tick_params(axis='both', which='major', labelsize=7)
ax.set_title('Sine wave advection', fontsize=9)

ax.set_xlabel('Grid resolution')
ax.set_ylabel(r'$L_1$' + ' Error')



fig.savefig('./sine_dtRedctn.eps', bbox_inches='tight', padding=0)
```
